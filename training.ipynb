```python
def convert(object):
    object=ast.literal_eval(object)
    tag_list=[]
    for dict_obj in object:
        tag_list.append(dict_obj['name'])
    return tag_list
```

```python
def convert3cast(object):
    c=0;
    object=ast.literal_eval(object)
    tag_list=[]
    for dict_obj in object:
        tag_list.append(dict_obj['name'])
        c=c+1
        if c==3:
            return tag_list
    return tag_list
```

The CountVectorizer class in sklearn.feature_extraction.text is used for converting a collection of text documents into a matrix of token (word) counts. It transforms text into a format that can be used for machine learning models, where the text data is represented as vectors of word frequencies.

```python
# Parameters:
# max_features: Limits the number of words in the vocabulary.
# stop_words: Removes common stop words like "and," "the," etc.
# ngram_range: Allows you to specify if you want to consider not just single words but also combinations of words (n-grams).
```

```python
def my_stem(txt):
    y=[]
    for i in txt.split():
        y.append(ps.stem(i))
    return " ".join(y)
   
```

```python
import sklearn
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tag']).toarray()
```

```python
features=cv.get_feature_names_out()
```

```python
for i in features:
    print(i)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarities=cosine_similarity(vectors)
similarities.shape
```

```python
def recommend_movied(movie):
    movie_index=new_df[new_df['title']==movie].index[0]
    distances=similarities[movie_index]
    list_tuple_index_similarity=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])
    top_5_movies=list_tuple_index_similarity[1:6]
    for movie_rec in top_5_movies:

        print(new_df.iloc[movie_rec[0]].title)
```

```python
recommend_movied('Batman Begins')
```
